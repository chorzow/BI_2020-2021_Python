```python
import random
import matplotlib.pyplot as plt

def carpet_plot(p):
    p_x = [x for (x,y) in p]
    p_y = [y for (x,y) in p]
    plt.title('Sierpinski carpet')
    plt.plot(p_x, p_y, '.')
    plt.show()

def carpet(iter_num):
    v = [(0.0, 0.0), (0.0, 0.5), (0.0, 1), (0.5, 1), (1, 1), (1, 0.5), (1, 0.0), (0.5, 0)]
    p = []
    x, y = random.choice(v)
    for _ in range(iter_num):
        new_x, new_y = random.choice(v)
        x = ((2 * new_x) + x) / 3.0
        y = ((2 * new_y) + y) / 3.0
        p.append((x, y))
    carpet_plot(p)

iter_num = 5000

carpet(iter_num)
```
