```python
import numpy as np 
import matplotlib.pyplot as plt
import random 

n = 10000  # n is the number of steps
x = np.zeros(n) # x and y store the coordinates of the position 
y = np.zeros(n)

direction = ['u', 'd', 'l', 'r'] # directions

for i in range(1, n): 
    step = random.choice(direction) # randomly choose direction
    
    if step == "r": # update the direction
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step == "l": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step == "u": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1

plt.title("2D Random Walk")
plt.plot(x, y)
plt.show()
```
