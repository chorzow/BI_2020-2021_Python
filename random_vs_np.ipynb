```python
import random
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def measure_time(module, k=1000):
    """
    Measure time of generation of k numbers for two libraries: random or numpy.
    module (required): library used to generate numbers. Can be either 'np' (stands for numpy) or 'random'.
    k: how many numbers to generate. Default is 1000.
    """
    timeline = []
    for i in range(k):
        time_begin = time.perf_counter()
        for _ in range(0, i):
            if module == 'random':
                random.random()
            elif module == 'np':
                np.random.random()
            else:
                raise ValueError(f"Error: argument 'module' can be 'random' or 'np', not {module}")
        time_end = time.perf_counter()
        timeline.append(time_end - time_begin)
    return timeline
```

```python
# example
k = 100
random_time = measure_time('np', k=k)
np_time = measure_time('random', k=k)

fig,ax = plt.subplots()
plt.xlabel('Amount of generated numbers')
plt.ylabel('Time')
plt.title('Calculation time of random numbers')
ax.plot(range(k), random_time, label = 'random')
ax.plot(range(k), np_time, label = 'numpy')
leg = ax.legend()
plt.show()
```
